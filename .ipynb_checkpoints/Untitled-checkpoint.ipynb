```python
from sklearn.preprocessing import StandardScaler
```

```python
X_data = data.drop('target',axis=1)
scaler = StandardScaler()
scaler = scaler.fit(X_data)
X_scaled_data = scaler.transform(X_data)
X_data = pd.DataFrame(X_scaled_data,columns=X_data.columns)
```

```python
X_data.describe()
```

```python
sc = StandardScaler()
train_std=sc.fit_transform(X_data)
train_std.shape
train_num=pd.DataFrame(train_std,columns=X_data.columns)
train_num.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.33, random_state=42)
```

```python
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
classification_report(y_pred,y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_train, test_size=0.33, random_state=42)
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
confusion_matrix(y_pred,y_test)
```

```python
classification_report(y_pred,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit,GridSearchCV
from sklearn.metrics import f1_score,make_scorer
```

```python
params={
    'gamma':[0.1,0.01,0,0.001,0.0001],
    'C':[0.1,10,100,1000],
    'kernel':['linear','rbf']
}
grid=GridSearchCV(clf,param_grid=params,scoring=make_scorer(f1_score))
```

```python
grid_clf=grid.fit(X_train,y_train)
```

```python
grid_clf.best_estimator_
```

```python
y_grid_pred=grid_clf.predict(X_test)
```

```python
confusion_matrix(y_grid_pred,y_test)
```
